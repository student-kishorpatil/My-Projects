```python
for i in X1.columns:                                                                                            # Outlier detection
    if(X1[i].dtype=="int64")|(X1[i].dtype=="float64"):
        print(i)
        Q1=X1[[i]].quantile(0.25)
        Q3=X1[[i]].quantile(0.75)
        Q2=X1[[i]].quantile(0.5)
        print('Q1=',Q1[0])
        print('Q2=',Q2[0])
        print('Q3=',Q3[0])
        IQR=Q3[0]-Q1[0]
        print('IQR',IQR)
        upp=Q3[0]+1.5*IQR
        low=Q1[0]-1.5*IQR
        print('UPP',upp)
        print('LOW',low)
        sns.boxplot(X1[[i]])
        plt.show()
        x=X1.loc[(X1[i]>=low)&(X1[i]<=upp)]
        sns.boxplot(X1[[i]])
        plt.show()
        
    
    
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
L=LogisticRegression()
from sklearn.model_selection import GridSearchCV
params={"C":[0.1,0.2,0.02,0.001],"penalty":["l1","l2"]}
G=GridSearchCV(L,param_grid=params,scoring="accuracy",cv=7)
```

```python
L.fit(x_train,y_train)
```

```python
G.fit(x_train,y_train)
```

```python
G.best_params_
```

```python
model=G.best_estimator_
```

```python
model
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(pred,y_test)                                                                        # confusion matrix
```

```python
accuracy_score(pred,y_test)                                                                            #accuracy_score
```

```python
print(classification_report(pred,y_test))                                                                #classification_report
```

```python
model.score(x_train,y_train)                                                                                  # training data accuracy
```

```python
model.score(x_test,y_test)                                                                                    # testing data accuracy
```
